подключаем библиотеки

```python
import json
import os
import requests
from dotenv import load_dotenv
```

ввод фильма и API-ключа


```python
load_dotenv()

title = input("Введите фильм для поиска: ").strip()
api_key = (os.getenv("OMDB_API_KEY") or "").strip()

if not title:
    raise ValueError("Название фильма не может быть пустым")

if not api_key or api_key == "ваш_ключ":
    raise ValueError("Укажите реальный OMDB_API_KEY в .env")
```

отправляем запрос

```python
try:
    response = requests.get(
        "https://www.omdbapi.com/",
        params={
            "t": title,
            "apikey": api_key,
        },
        timeout=10,
    )
    response.raise_for_status()
except requests.exceptions.RequestException as exc:
    raise RuntimeError("Ошибка запроса к OMDb API") from exc
```

превращаем ответ в JSON

```python
try:
    json_data = response.json()
except ValueError as exc:
    raise ValueError("OMDb вернул невалидный JSON") from exc

if json_data.get("Response") == "False":
    raise ValueError(f"OMDb API error: {json_data.get('Error', 'unknown error')}")

print(json.dumps(json_data, ensure_ascii=False, indent=2))
```

достаем нужные поля

```python
movie_title = json_data.get("Title")
year = json_data.get("Year")
genre = json_data.get("Genre")
director = json_data.get("Director")
imdb_rating = json_data.get("imdbRating")
print(f"Название: {movie_title}")
print(f"Год: {year}")
print(f"Жанр: {genre}")
print(f"Режиссёр: {director}")
print(f"Рейтинг IMDb: {imdb_rating}")
```

изменения в ветви 

```python
# вывод актеров
actors = json_data.get("Actors")
print(f"Актёры: {actors}")

# красивый итоговый вывод
print(f"Фильм '{movie_title}' ({year}) — жанр: {genre}. Режиссёр: {director}.")
```
